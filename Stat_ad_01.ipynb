
1. What is a random variable in probability theory?  
   - A variable that represents numerical outcomes of a random experiment.

2. What are the types of random variables?  
   - Discrete: Takes specific values (e.g., number of heads in coin flips).  
   - Continuous: Takes infinite values within a range (e.g., height, temperature).  

3. What is the difference between discrete and continuous distributions?  
   - Discrete: Uses a Probability Mass Function (PMF), sum of probabilities = 1.  
   - Continuous: Uses a Probability Density Function (PDF), integrates to 1.

4. What are probability distribution functions (PDF)?  
   - Functions describing the probability of different values in a continuous distribution.

5. How do cumulative distribution functions (CDF) differ from PDFs?  
   - PDF: Probability at a specific point.  
   - CDF: Cumulative probability up to a value.

6. What is a discrete uniform distribution?  
   - A distribution where all outcomes have equal probability (e.g., rolling a fair die).

7. What are the key properties of a Bernoulli distribution?  
   - Models a single trial with success/failure.  
   - Mean = \( p \), Variance = \( p(1 - p) \).

8. What is the binomial distribution, and how is it used in probability?  
   - Models multiple Bernoulli trials.  
   - Used for success/failure counts (e.g., coin flips).

9. What is the Poisson distribution and where is it applied?  
   - Models rare events in a fixed interval.  
   - Used in traffic flow, call arrivals, etc.

10. What is a continuous uniform distribution?  
    - A range \([a, b]\) where all values are equally likely.

11. What are the characteristics of a normal distribution?  
    - Bell-shaped curve.  
    - Mean = Median = Mode.  
    - Follows the 68-95-99.7 rule.

12. What is the standard normal distribution, and why is it important?  
    - Mean = 0, Variance = 1.  
    - Used for Z-score standardization.

13. What is the Central Limit Theorem (CLT), and why is it critical?  
    - Regardless of original distribution, sample means approximate a normal distribution as sample size grows.

14. How does the CLT relate to the normal distribution?  
    - CLT explains why normal distributions commonly appear in real-world scenarios.

15. What is the application of Z statistics in hypothesis testing?  
    - Helps compare sample data to population parameters.

16. How do you calculate a Z-score, and what does it represent?  
    \[
    Z = \frac{X - \mu}{\sigma}
    \]  
    Measures how far a value is from the mean.

17. What are point estimates and interval estimates in statistics?  
    - Point estimate: Single value estimation.  
    - Interval estimate: Range with confidence level.

18. What is the significance of confidence intervals in statistical analysis?  
    - Provides a range where the population parameter likely lies.

19. What is the relationship between a Z-score and a confidence interval?  
    - Z-score determines critical values for confidence intervals.

20. How are Z-scores used to compare different distributions?  
    - Standardizes values across different distributions.

21. What are the assumptions for applying the Central Limit Theorem?  
    - Independence of samples.  
    - Sufficiently large sample size.

22. What is the concept of expected value in a probability distribution?  
    \[
    E(X) = sum X P(X)
    \]  
    Represents the long-run average outcome.

23. How does a probability distribution relate to the expected outcome of a random variable?  
    - Defines the likelihood of different values occurring over multiple trials.
